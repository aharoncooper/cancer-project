```python
import pickle
with open('final_df.pkl', 'rb') as file:
    final_df = pickle.load(file)
```

```python
df=final_df
df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn.metrics as metrics 
```

## normalization

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_exclude = ['diagnosis', 'id']

data_to_normalize = df.drop(columns=columns_to_exclude)
data_to_normalize = data_to_normalize.select_dtypes(include=['float64', 'int64'])  

scaler = MinMaxScaler()
normalized_values = scaler.fit_transform(data_to_normalize)

normalized_df = pd.DataFrame(normalized_values, columns=data_to_normalize.columns)

normalized_df['diagnosis'] = df['diagnosis'].values
normalized_df['id'] = df['id'].values
```

```python
df=normalized_df
df
```

```python
df.set_index('id', inplace= True)
df
```

```python
df.to_pickle('dataframe.pkl')
```

```python
def classificationMetrics(y, yhat):
    prf1 = metrics.precision_recall_fscore_support(y,yhat)
    res = {'Accuracy': metrics.accuracy_score(y,yhat),
           'Precision':prf1[0][1],
           'Recall': prf1[1][1],
           'f1-score': prf1[2][1],
           'Log-loss': metrics.log_loss(y,yhat),
           'AUC': metrics.roc_auc_score(y,yhat)
          }
    return res
```

```python
X = df.loc[:,df.columns != 'diagnosis']
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import xgboost as xgb
```

```python
models_list = pd.DataFrame()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
mod1 = LogisticRegression(max_iter=10000)
mod1.fit(X_train,y_train)

pred1 = mod1.predict(X_test)
```

```python
mod1 = LogisticRegression(random_state=1, max_iter=10000)
mod1.fit(X,y)

pred1 = mod1.predict(X)
```

```python
classificationMetrics(y, pred1)
```

```python
model_dict = {'model': "Logistic Regression"}
new_row = pd.DataFrame([{**model_dict, **classificationMetrics(y, pred1)}])
models_list = pd.concat([models_list, new_row], ignore_index=True)
models_list
```

```python
pd.crosstab(y, pred1)
```
